# **Data Preparation**

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("student.csv")
```

```python
df
```

# ลบแถวที่คอลัมน์ 'ชื่อ' มีค่าเป็น NaN

```python
df = df.dropna(subset=['ชื่อ'])
```

```python
df
```

# ลบแถวคอลัมน์ 'อายุ' ที่มีค่าผิดปกติ โดยที่ค่า NaN ยังอยู่

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df['อายุ'] >= Q1 - 1.5 * IQR) & (df['อายุ'] <= Q3 + 1.5 * IQR)) | df['อายุ'].isna()

df = df[outliers]
```

```python
df
```

# ใส่ค่าเฉลี่ยของอายุ ในคอลัมน์ 'อายุ' ที่เป็นค่า NaN

```python
mean_age = df['อายุ'].mean()
df['อายุ'].fillna(mean_age, inplace=True)
df
```

# แปลงค่าทศนิยมที่คอลัมน์ 'อายุ' ให้เป็นจำนวนเต็ม

```python
df['อายุ'] = df['อายุ'].astype(int)
df
```

# ทำคอลัมน์ 'เพศ' ให้เป็นค่ามาตรฐาน

```python
def standardize_faculty(name):
    if name in ['ชาย', 'ช', 'M', 'Male']:
        return 'ชาย'
    elif name in ['หญิง', 'ญ', 'F', 'Female']:
        return 'หญิง'
    else:
        return name

df['เพศ'] = df['เพศ'].apply(standardize_faculty)
```

# ใส่ข้อมูลในคอลัมน์ 'เพศ' ที่เป็นค่า NaN โดยเติมค่าด้วยค่าที่พบบ่อยที่สุด

```python
df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
df
```

# ทำคอลัมน์ 'คณะ' ให้เป็นค่ามาตรฐาน



```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวฯ', 'วิศวกรรม', 'วิศวะ']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name

df['คณะ'] = df['คณะ'].apply(standardize_faculty)
df
```

# ใส่ข้อมูลในคอลัมน์ 'คณะ' ที่เป็นค่า NaN โดยเติมค่าด้วยค่าที่พบบ่อยที่สุด

```python
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
df
```

# ทำคอลัมน์ 'GPA' ที่มีค่าผิดปกติให้เป็นค่า NaN

```python
df.loc[((df['GPA'] < 0) | (df['GPA'] > 4.00)) & df['อายุ'].notna(), 'GPA'] = np.nan
df
```

# ใส่ค่าเฉลี่ยของGPA ในคอลัมน์ 'GPA' ที่เป็นค่า NaN

```python
mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)
df['GPA'] = df['GPA'].round(2)
df
```

# ใส่ค่าเฉลี่ยของข้อมูล ในคอลัมน์ (ส่วนสูง_cm,	น้ำหนัก_kg,	ชั่วโมงเรียน_ต่อสัปดาห์) ที่มีค่าเป็น NaN

```python
mean_high = df['ส่วนสูง_cm'].mean()
df['ส่วนสูง_cm'].fillna(mean_high, inplace=True)
df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].round(2)


mean_weight = df['น้ำหนัก_kg'].mean()
df['น้ำหนัก_kg'].fillna(mean_weight, inplace=True)
df['น้ำหนัก_kg'] = df['น้ำหนัก_kg'].round(1)

mean_lern = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_lern, inplace=True)
df['ชั่วโมงเรียน_ต่อสัปดาห์'] = df['ชั่วโมงเรียน_ต่อสัปดาห์'].astype(int)

df
```

# **Feature Engineering**

```python
data = df
data
```

# แปลงข้อมูลของคอลัมน์ 'คณะ' เป็นตัวเลข

```python
data_encoded = pd.get_dummies(data, columns=['คณะ']) * 1
data_encoded
```

# การสร้าง Feature จากเกรด GPA

```python
g_data = data.copy()

g_data['ประเภทนักเรียน'] = pd.cut(g_data['GPA'],
                                  bins=[0, 3.25, 4.00],
                                  labels=['เรียนปกติ', 'เรียนเด่น'])

g_data
```

# การรวม Features สร้าง BMI จากส่วนสูงและน้ำหนัก และสร้าง สุขภาพ

```python
student_health = data.copy()


student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

student_health
```
